```python
#A1_a
def summation_unit(x0,x1,x2,w0,w1,w2): #summation unit
  return (x0*w0)+(x1*w1)+(x2*w2)
```

```python
#A1_b
def step_function(x):#step_function
  if x>0:
    return 1
  else:
    return 0
```

```python
#A1_b
def bipolar_function(x):#bipolar_function
  if x>0:
    return 1
  elif x == 0:
    return 0
  else:
    return -1
```

```python
#A1_b
def sigmoid_function(x):#sigmoid_function
  return (1/(1+(math.e**-x)))
```

```python
#A1_b
def relu(x):#relu
  if x>0:
    return x
  else:
    return 0
    
```

```python
#A2
# Initial weights for the perceptron
w0 = 10    # Weight for bias term
w1 = 0.2   # Weight for input A
w2 = -0.75 # Weight for input B

# Learning rate controls the step size during weight updates
learning_rate = 0.5

# Initialize epoch count and lists to track progress
epoch_count = 0
epoches = []             # To store the number of epochs
sum_square_error = []    # To store the sum of squared errors for each epoch

# Training loop: continue for a maximum of 100 epochs
while epoch_count < 100:
    error = 0  # Reset error for this epoch
    
    # Iterate over each row (data point) in the dataset (and_table)
    for index, row in and_table.iterrows():
        
        # Predict the output using the step function and summation unit
        if step_function(summation_unit(row['one'], row['A'], row['B'], w0, w1, w2)) != row['result']:
            
            # Calculate the error: difference between actual and predicted output
            e = row['result'] - step_function(summation_unit(row['one'], row['A'], row['B'], w0, w1, w2))
            
            # Update the weights using the perceptron learning rule
            w0 += (learning_rate * e * row['one'])  # Update weight for bias
            w1 += (learning_rate * e * row['A'])    # Update weight for input A
            w2 += (learning_rate * e * row['B'])    # Update weight for input B
            
            # Accumulate squared error for this epoch
            error += e**2
    
    # Record the epoch count and total squared error
    epoches.append(epoch_count)
    epoch_count += 1
    sum_square_error.append(error)
    
    # Early stopping condition: stop if error is very small
    if error < 0.002:
        print("Epoch count: ", epoch_count)
        break

# Plot the sum of squared errors across epochs to visualize learning progress
plt.plot(epoches, sum_square_error)
plt.show()

# Print final weight values after training
print("Weights ==> w0: ", w0, " w1: ", w1, " w2: ", w2)

# Print the predicted and actual output for each data point in the dataset
for index, row in and_table.iterrows():
    print("Calculated output: ", step_function(summation_unit(row['one'], row['A'], row['B'], w0, w1, w2)))
    print("Real output: ", row['result'])
```

```python
#A8
# Initial weights for input-to-hidden layer connections (V) and hidden-to-output layer connections (W)
V = [[1, -1], [1, 1]]  # Weights from input (A, B) to hidden layer (2 neurons)
W = [1, 1]             # Weights from hidden layer (2 neurons) to output

# Set learning rate for weight updates
learning_rate = 0.05

# Initialize iteration counter and error tracker
iter = 0
error = 0
errors = []  # To track error over iterations

# Main training loop: Run for 1000 iterations or until error is sufficiently small
while iter < 1000 and error > 0.02 or error == 0:
    error = 0  # Reset error for the current iteration
    
    # Iterate through each row in the AND table dataset
    for index, row in And_table.iterrows():
        
        # Forward pass: calculate input and output for hidden layer
        input_hidden = []  # Input values for hidden layer neurons
        hidden_out = []    # Output values for hidden layer neurons (after activation)
        
        # Input to hidden neuron 1
        input_hidden.append(row['A'] * V[0][0] + row['B'] * V[1][0])
        # Input to hidden neuron 2
        input_hidden.append(row['A'] * V[0][1] + row['B'] * V[1][1])
        
        # Apply sigmoid activation function to inputs of hidden neurons
        hidden_out.append(sigmoid_function(input_hidden[0]))
        hidden_out.append(sigmoid_function(input_hidden[1]))
        
        # Output layer: sum weighted outputs of hidden layer neurons and apply sigmoid activation
        output = sigmoid_function(hidden_out[0] * W[0] + hidden_out[1] * W[1])
        
        # Calculate the error (difference between actual and predicted output)
        ERROR = row['Output'] - output
        
        # Backpropagation: Calculate deltas for output and hidden layers
        delta1 = ERROR * output * (1 - output)  # Delta for output layer
        
        # Deltas for hidden layer neurons
        delta2 = [
            delta1 * W[0] * hidden_out[0] * (1 - hidden_out[0]),  # Delta for hidden neuron 1
            delta1 * W[1] * hidden_out[1] * (1 - hidden_out[1])   # Delta for hidden neuron 2
        ]
        
        # Update weights between hidden layer and output
        for i in range(len(W)):
            W[i] += learning_rate * delta1 * hidden_out[i]
        
        # Update weights between input and hidden layer
        V[0][0] += learning_rate * delta2[0] * row['A']
        V[0][1] += learning_rate * delta2[1] * row['A']
        V[1][0] += learning_rate * delta2[0] * row['B']
        V[1][1] += learning_rate * delta2[1] * row['B']
        
        # Accumulate squared error for this iteration
        error += ERROR ** 2
    
    # Record error for plotting
    errors.append(error)
    
    # Increment iteration counter
    iter += 1

# Plot the error over iterations to visualize learning progress
plt.plot(errors)
plt.xlabel('Iterations')
plt.ylabel('Error')
plt.title('Training Error Across Iterations')
plt.show()
```
